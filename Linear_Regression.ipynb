```python
#Basic from scratch Linear Regression classifier implementation in Python


#Dependencies
import math
import numpy as np
import pandas as pd
```

```python
#First we define the loss function


def mean_squared_error(output_actual: list, output_predicted: list) -> float:
    '''
    This function takes as input the actual output and the predicted output from the model and uses the mean squared error (MSE) function to find the loss.
    The MSE is found by first taking the models output for a certain value, subtracts the actual output. Then we square that.
    Then it finds the average of all these squares (sum of all squares divided by the amount of points).
    '''
    #We initialise the MSE value
    MSE_value: float = 0.0
    #This tells us how many inputs we have to check (TO DO: Make sure input lists are same size error)

    no_points: int = len(output_actual)
    i = 0
    #Here we find the sum of the squares and add it to the MSE value, loop for each.
    while i < no_points:
        difference = output_predicted[i]-output_actual[i]
        MSE_value += (difference**2)
        i += 1

    #Here we divide the final MSE sum by the number of outputs and return the value
    MSE_value = MSE_value / float(len(output_actual))
    
    return MSE_value

def derive_rename_idk(output_actual: list, output_predicted: list, input_actual: list) -> float:
    '''
    -2 * (actual_y - (bias + w1*x1))
    '''
    #We initialise the MSE value

    derivative_bias: float = float()
    derivative_weight: float = float()

    #This tells us how many inputs we have to check (TO DO: Make sure input lists are same size error)

    i = 0
    no_points: int = len(output_actual)
    #Here we find the sum of the squares and add it to the MSE value, loop for each.

    while i < no_points:
        derivative_bias += (-2.0*(output_actual[i]-output_predicted[i]))
        derivative_weight += input_actual[i]*(-2.0*(output_actual[i]-output_predicted[i]))
        i+=1
    derivative_bias = derivative_bias / float(len(output_actual))
    derivative_weight = derivative_weight / float(len(output_actual))

    return [derivative_weight, derivative_bias]


def fit(df_x, df_Y,epochs: int = 100000, learning_rate: float = 0.00000001) -> list:
    #Base hypothesis, 0 for every value + bias
    
    p_x: list = [0.0,0.0]
    p_Y = list()
    
    df_x = df_x.tolist()
    df_Y = df_Y.tolist()

    epochs_count = epochs
    while epochs_count > 0:
        p_Y: list = make_prediction(df_x, p_x)    

        derivative = derive_rename_idk(df_Y, p_Y, df_x)

        new_weight = p_x[0]-learning_rate*derivative[0]

        new_bias = p_x[1]-learning_rate*derivative[1]

        p_x = [new_weight, new_bias]

        epochs_count -= 1
    print(f"{epochs} epochs, learning rate: {learning_rate}, weight: {p_x[0]}, bias {p_x[1]}")


    return p_x

def RMSE(df_y, p_y):

    RMSE = math.sqrt((mean_squared_error(df_y, p_y)))
    return RMSE

def make_prediction(df_x_test, p_x):
    
    p_Y = list()
    for value in df_x_test:
        p_Y.append(value*p_x[0]+p_x[1])

    return p_Y

def predict(df_x, df_y, p_x):

    p_Y = make_prediction(df_x, p_x)

    return RMSE(df_y, p_Y)





```

```python
#https://www.kaggle.com/datasets/andonians/random-linear-regression was used

df_lr = pd.read_csv("./train.csv")

print(df_lr)


df_lr = df_lr.dropna().reset_index(drop=True)

df_lr_x = df_lr.x

df_lr_y = df_lr.y


print(df_lr_x)

test = pd.read_csv("./test.csv")


test = test.dropna().reset_index(drop=True)

test_x = test.x

test_y = test.y

p_x = (fit(df_lr_x, df_lr_y))
prediction = predict(test_x, test_y, p_x)

print(f"RMSE of predicted data and actual data: {prediction}")


#Could add a train/test split method for this model at some point.



```
